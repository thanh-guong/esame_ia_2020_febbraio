## Spearman correlation

Utilizzando la correlazione di Spearman è possibile costruire una matrice di correlazione per avere un'idea di come siano correlate le features tra di loro.

Come definito su [wikipedia](https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient):
> Intuitively, the Spearman correlation between two variables will be high when observations have a similar (or identical for a correlation of 1) rank (i.e. relative position label of the observations within the variable: 1st, 2nd, 3rd, etc.) between the two variables, and low when observations have a dissimilar (or fully opposed for a correlation of −1) rank between the two variables. 

```python
from scipy import stats

correlation_matrix = []

# per ogni nome delle colonne del DataFrame 
for i in (dataset.columns):
    # sto costruendo una matrice, e questa è l'i-esima colonna di questa matrice
    column_i = []
    
    # per ogni nome delle colonne del DataFrame
    for j in (dataset.columns):
       
        # calcola la correlazione tra l'i-esimo vettore delle features, e il j-esimo vettore delle features
        correlation = correlation = stats.spearmanr(dataset[i], dataset[j])
        # aggiungi il coefficiente di correlazione, nel vettore colonna che rappresenta l'i-esima colonna
        column_i.append(correlation.correlation)
        
    # aggiungi il vettore colonna alla matrice di correlazione
    correlation_matrix.append(column_i)
```

```python
import numpy as np

# preparing to be printed out, just aesthetics using pandas DataFrame
correlation_matrix = np.matrix(correlation_matrix)
correlation_matrix = pd.DataFrame(correlation_matrix, columns = dataset.columns)
correlation_matrix.index = dataset.columns
```

```python
correlation_matrix
```

```python
strongly_correlated_features = set()

correlation_matrix['status']['status']

for column_name in dataset.columns:
    for row_name in dataset.columns:
        if row_name == column_name:
            break

        if abs(correlation_matrix[row_name][column_name]) > spearman_treshold:
            strongly_correlated_features.add(str(row_name) + " - " + str(column_name))

if len(strongly_correlated_features):
    print("Feature couples with Spearman correlation above " + str(spearman_treshold) + ":\n")
else:
    print("There isn't any couple of features with Spearman correlation above " + str(spearman_treshold))
for scf in strongly_correlated_features:
    print(scf)
```

# Data transformation

Qui vengono trasformati il dataset di training e di test:
- **Rimuovere features**
    - salary
- **Arrotondamento delle features**
    - ssc_p
    - hsc_p

Le motivazioni di queste scelte sono descritte nella sezione Data Analysis (sopra).

## Rimozione features

```python
%%capture
column_name = 'salary'
dataset.drop(columns=[column_name], inplace=True)
training_set.drop(columns=[column_name], inplace=True)
test_set.drop(columns=[column_name], inplace=True)
```

## Arrotondamento delle features

```python
column_names = ['ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p']

for column_name in column_names:
    dataset[column_name] = dataset[column_name].astype('int64')
    training_set[column_name] = training_set[column_name].astype('int64')
    test_set[column_name] = test_set[column_name].astype('int64')
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder

column_names = ['gender', 'ssc_b', 'hsc_b', 'hsc_s', 'degree_t', 'workex', 'specialisation']

encoder = LabelEncoder()

for column_name in column_names:
    dataset[column_name] = encoder.fit_transform(dataset[column_name])
    training_set[column_name] = encoder.fit_transform(training_set[column_name])
    test_set[column_name] = encoder.fit_transform(test_set[column_name])
```

## Ricostruzione di y_train e y_test

Producendo files csv in output dal file di preparazione, si sono persi i vettori delle coppie (ID, Label), che vanno dati in pasto alla funzione fit della LogisticRegression.

```python
# x_test reconstruction

y_train = {}

for index, training_datapoint in training_set.iterrows():
    training_datapoint_label = training_datapoint[label_column_name]
    y_train[index] = training_datapoint_label

y_train = pd.Series(y_train, name=label_column_name)

# y_test reconstruction    

y_test = {}

for index, test_datapoint in test_set.iterrows():
    test_datapoint_label = test_datapoint[label_column_name]
    y_test[index] = test_datapoint_label
    
y_test = pd.Series(y_test, name=label_column_name)
```

## Label dropping

E' necessario droppare le colonne contenenti le labels, altrimenti il modello viene addestrato sulla base di queste etichette (non si può usare il modello, servirebbe l'etichetta per usarlo).

# Training phase

Questa è la fase in cui addestro il classificatore con regressione logistica.

L'algoritmo scelto è **liblinear** in quanto il numero di datapoints non è così esteso, e sulla [documentazione](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) di sklearn c'è scritto 
> For small datasets, ‘liblinear’ is a good choice, whereas ‘sag’ and ‘saga’ are faster for large ones.



```python
from sklearn.linear_model import LogisticRegression

# istanziamento dell'oggetto per il modello
lr = LogisticRegression(solver='liblinear', penalty="l1")

# training
lr = lr.fit(training_set, y_train)
```

# Evaluation phase

Qui vengono testate le prestazioni del modello basandosi sui datapoints dedicati al testing.

Viene effettuata una predizione per ogni datapoint del test set, calcolata la matrice di confusione, generato il classification report di sklearn, e calcolato lo score (accuracy) della logistic regression sia per il training set, che per il test set.

Ogni dato raccolto nella procedura descritta qui sopra, conferma le elevate prestazioni del modello.

## Test

```python
# vettore contenente il risultato del classificatore quando viene passato in input il test set
pred = lr.predict(test_set)
```

## Confusion matrix

```python
from sklearn.metrics import confusion_matrix

#matrice di confusione
confusion_matrix = np.matrix(confusion_matrix(pred, y_test))
confusion_matrix = pd.DataFrame(confusion_matrix, columns = ["Predicted as Placed", "Predicted as Not Placed"])
confusion_matrix.index = ['Placed', 'Not Placed']
```

```python
confusion_matrix
```

## Classification report

```python
from sklearn.metrics import classification_report

print(classification_report(pred, y_test))
```

## Training error

```python
training_score = lr.score(training_set, y_train).round(2)
training_error = (1 - training_score).round(2)
print("Training error: " + str(training_error))
```

## Test error

```python
test_score = lr.score(test_set, y_test).round(2)
test_error = (1 - test_score).round(2)
print("Test error: " + str(test_error))
```

## Accuracy

```python
print("Accuracy: " + str(test_score))
```
