## Configurazione

Qui vengono salvati alcuni valori che vengono utilizzati spesso nel file, lo scopo è quello di evitare hard coding.

```python
dataset_full_path = "./"
dataset_filename = "Placement_Data_Full_Class.train.csv"

training_set_filename = "train.csv"
test_set_filename = "test.csv"

label_column_name = "status"

random_state = 11
```

# Data prepare

In questa fase viene letto il dataset, processato e splittato in due files: train.csv (contenente il dataset di training), e test.csv (contenente il dataset di test):
- Data import
- Data split
- Data export

## Data import

Viene letto il dataset in un pandas DataFrame.

```python
# pandas è una libreria per Data Analysis
import pandas as pd

dataset = pd.read_csv(dataset_filename)
```

## Data split

Qui viene effettuato lo split del dataset in training set e test set.

Utilizzando la funzione train_test_split della libreria sklearn, è possibile ottenere uno split dei dati in training set e test set, i dati prodotti sono:
- x_train = pandas DataFrame contenente i datapoints appartenenti al **training set**.
- y_train = pandas Series contenente le coppie (ID, Label) appartenenti al **training set**.
- x_test = pandas DataFrame contenente i datapoints appartenenti al **test set**.
- y_test = pandas Series contenente le coppie (ID, Label) appartenenti al **test set**.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(dataset, dataset[label_column_name], random_state = random_state)
```

## Data export

Come richiesto, il training set e il test set vengono esportati con nome train.csv e test.csv.

```python
x_train.to_csv(training_set_filename, index=0)
x_test.to_csv(test_set_filename, index=0)
```
